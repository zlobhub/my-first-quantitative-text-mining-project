```python

def clean_text(text):
    
    """
    Function to clean text with basic methods - HTML removal, whitespace correction, etc.
    """

    # Check if the text is not NaN
    if pd.isnull(text):
        return ""
    
    # Remove HTML content
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Remove newlines and adjacent spaces
    text = text.replace('\n', ' ').replace('\r', '').replace('\xa0', ' ')
    
    # Remove any remaining non-breaking spaces and other annoying characters
    text = text.replace(u'\xa0', u' ')
    
    # Remove excessive whitespace
    text = ' '.join(text.split())
    
    return text


# Apply the clean_text() function to the text column
data['text'] = data['text'].apply(clean_text)


# After cleaning, we need to drop rows with empty 'text' fields
initial_count = data.shape[0]
data = data[data['text'].str.strip() != ""]
final_count = data.shape[0]

# Return the number of dropped rows and sample cleaned texts to evaluate the cleaning
dropped_rows = initial_count - final_count
sample_cleaned_texts = data['text'].sample(5).tolist()

dropped_rows, sample_cleaned_texts
```

```python
# Extracting the ending portions of each text to identify common boilerplate phrases
endings = data['text'].apply(lambda x: ' '.join(x.split()[-100:]))

# Finding common phrases at the end of the texts which may indicate boilerplate content
endings_str = ' '.join(endings.tolist())
end_phrases = {}

# Split the consolidated endings by space to check for repetitive phrases
for phrase in endings_str.split('. '):
    if phrase in end_phrases:
        end_phrases[phrase] += 1
    else:
        end_phrases[phrase] = 1

# Identify phrases that occur most commonly
common_phrases = {k: v for k, v in sorted(end_phrases.items(), key=lambda item: item[1], reverse=True) if v > 1}
common_phrases
```

```python
# Define the boilerplate phrases to be removed
boilerplate_phrases = ["Kövesse az Indexet Facebookon is! Követem!"]

# Function to remove the identified boilerplate phrases
def remove_boilerplate(text, boilerplates):
    for phrase in boilerplates:
        text = text.replace(phrase, '')
    return text.strip()  # Remove leading/trailing whitespace

# Apply the boilerplate removal function to the cleaned text
data['text'] = data['text'].apply(lambda x: remove_boilerplate(x, boilerplate_phrases))

# Display some samples to verify if boilerplate text has been removed
sample_final_texts = data['text'].sample(5).tolist()

#sample_final_texts.rename(columns={'text': 'text_final'})


sample_final_texts
```

```python
# removing duplicates
data = data.drop_duplicates()


#order by date
data.sort_values(by='date')

#making a new column based on if it was before or after the resignation of the majority of journalists
#data[''] = []

# Selecting necessary columns to keep ('title_cleaned' and 'text_final') and dropping unnecessary ones
final_data = data[["date", "link","author","text"]]


# Save the cleaned data to a new CSV file
final_file_path = 'C:/Users/userű/Desktop/szakdoga/code/scrap_clean_newest/cleaned_putin_text.csv'
final_data.to_csv(final_file_path, index=False, sep=",")

final_file_path
```
